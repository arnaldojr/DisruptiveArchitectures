## Limpeza e Prepara√ß√£o de Dados

Na vida real, os dados raramente v√™m limpos e prontos para an√°lise. Vamos simular alguns problemas comuns e aprender a resolv√™-los.

### Problemas Comuns:
- **Valores faltantes** (NaN, null)
- **Dados duplicados**
- **Outliers** (valores extremos)
- **Inconsist√™ncias** nos dados

### Estrat√©gias de Tratamento:
1. **Remo√ß√£o**: Deletar linhas/colunas problem√°ticas
2. **Imputa√ß√£o**: Preencher com valores calculados

````python
# Simulando problemas nos dados para fins educativos
print("SIMULANDO PROBLEMAS NOS DADOS:")
print("=" * 40)

# Criando uma c√≥pia para experimentar
df_dirty = df.copy()

# Introduzindo valores faltantes aleatoriamente
np.random.seed(42)  # Para reprodutibilidade
for col in df_dirty.columns[:-1]:  # N√£o mexe na coluna species
    # Seleciona 5 posi√ß√µes aleat√≥rias para cada coluna
    missing_indices = np.random.choice(df_dirty.index, 5, replace=False)
    df_dirty.loc[missing_indices, col] = np.nan

print("Valores faltantes inseridos!")
print(f"Total de valores faltantes: {df_dirty.isnull().sum().sum()}")
````

## Estrat√©gias de Limpeza

Vamos explorar diferentes abordagens para lidar com valores faltantes:

### Estrat√©gia 1: Remo√ß√£o com dropna()
- **Vantagem**: Simples e r√°pida
- **Desvantagem**: Perde informa√ß√£o
- **Quando usar**: Poucos valores faltantes (< 5%)




````python
# Estrat√©gia 1: Remo√ß√£o de linhas com valores faltantes

df_dropped = df_dirty.dropna()

display(df_dirty.info(),df_dropped.info())
````

### Estrat√©gia 2: Imputa√ß√£o com fillna()
- **M√©dia**: Para dados num√©ricos sem outliers
- **Mediana**: Para dados com outliers
- **Moda**: Para dados categ√≥ricos
- **Interpola√ß√£o**: Para s√©ries temporais


````python
# Estrat√©gia 2: Imputa√ß√£o com m√©dia


df_imputed = df_dirty.copy()
numeric_columns = df_imputed.select_dtypes(include=[np.number]).columns

# Calcula m√©dias
mean_values = df_imputed[numeric_columns].mean()

# Conta valores ausentes
missing_counts = df_imputed[numeric_columns].isnull().sum()

# Preenche os valores
df_imputed.fillna(mean_values, inplace=True)

# Imprime resumo
for col in numeric_columns:
    print(f"{col}: {missing_counts[col]} valores preenchidos com {mean_values[col]:.2f}")

# Verifica resultado
print(f"\nValores faltantes ap√≥s imputa√ß√£o: {df_imputed.isnull().sum().sum()}")

display(df_dirty.info(), df_imputed.info())
````

###  Scatter Plots: Descobrindo Relacionamentos

Os **Scatter Plots** revelam relacionamentos entre vari√°veis. S√£o fundamentais para entender correla√ß√µes e padr√µes.

### Padr√µes a observar:
- **Linear**: Pontos seguem uma linha reta
- **Curvil√≠neo**: Rela√ß√£o n√£o-linear
- **Clusters**: Grupos distintos de pontos
- **Outliers**: Pontos isolados


````python
# Scatter plot b√°sico
# Grafico de dispers√£o

plt.scatter(df['sepal_length'], df['petal_length'])

plt.title('Sepal Length vs Petal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.show()
````

````python
# Os mesmos dados mas agora cada classe de uma cor diferente

colors = {'Iris-setosa':'red', 'Iris-versicolor':'blue', 'Iris-virginica':'green'}

plt.scatter(df['sepal_length'], df['petal_length'], c=df['species'].map(colors), label=colors)

plt.title('Sepal Length vs Petal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.legend(handles=[plt.Line2D([0], [0], marker='o', color='w', markerfacecolor=color, markersize=10) for color in colors.values()], labels=colors.keys())
plt.show()
````

````python
# Usando seaborn para um pair plot profissional
import seaborn as sns

plt.figure(figsize=(12, 10))
sns.pairplot(df, hue='species', height=2.5, aspect=1.2)

````

## An√°lise de Correla√ß√£o

A **correla√ß√£o** mede a for√ßa e dire√ß√£o do relacionamento linear entre vari√°veis.

### üìè Escala de Interpreta√ß√£o:
| Valor | Interpreta√ß√£o |
|-------|---------------|
| 0.9 a 1.0 | Correla√ß√£o muito forte |
| 0.7 a 0.9 | Correla√ß√£o forte |
| 0.5 a 0.7 | Correla√ß√£o moderada |
| 0.3 a 0.5 | Correla√ß√£o fraca |
| 0.0 a 0.3 | Correla√ß√£o desprez√≠vel |

### Importante:

**Correla√ß√£o ‚â† Causalidade**
- Uma correla√ß√£o alta n√£o significa que uma vari√°vel causa a outra
- Pode haver uma terceira vari√°vel influenciando ambas


````python
cols = ['sepal_length', 'sepal_width', 'petal_length','petal_width']
corr_matx = df[cols].corr()

heatmap = sns.heatmap(corr_matx,cbar=True,annot=True,square=True,fmt='.2f',annot_kws={'size': 15},yticklabels=cols,xticklabels=cols,cmap='Dark2')
````
