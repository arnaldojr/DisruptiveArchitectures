## O que é Análise Exploratória de Dados (EDA)?

A **Análise Exploratória de Dados** é como fazer uma "investigação" dos seus dados antes de aplicar qualquer modelo ou algoritmo. É o momento em que você:

- **Explora** os dados para entender sua estrutura
- **Limpa** e prepara os dados para análise
- **Visualiza** padrões e tendências
- **Gera** insights e hipóteses

### Analogia: O Detetive de Dados

Imagine que você é um detetive investigando um mistério. Os dados são suas pistas, e a EDA é o processo de:
1. Catalogar todas as evidências (conhecer os dados)
2. Procurar pistas falsas ou danificadas (limpeza)
3. Encontrar padrões suspeitos (visualizações)
4. Formular teorias sobre o caso (insights)

## Conhecendo o Dataset Iris

### Sobre o Dataset

O **Iris Dataset** é um dos conjuntos de dados mais famosos na ciência de dados, criado pelo biólogo Ronald Fisher em 1936. Ele contém medições de diferentes espécies de flores íris.

### Características (Features):
1. **sepal_length**: Comprimento da sépala (cm)
2. **sepal_width**: Largura da sépala (cm)  
3. **petal_length**: Comprimento da pétala (cm)
4. **petal_width**: Largura da pétala (cm)
5. **species**: Espécie da flor (classe alvo)

<img src="https://s3.amazonaws.com/assets.datacamp.com/blog_assets/Machine+Learning+R/iris-machinelearning.png" alt="Anatomia da Flor Iris" width="500">

### Espécies:
- **Iris Setosa**: Geralmente menor, pétalas mais estreitas

````python
# Carregando o dataset Iris
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"

# Definindo os nomes das colunas conforme documentação
column_names = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']

# Carregando os dados com os nomes corretos das colunas
df = pd.read_csv(url, header=None, names=column_names)

print(f"📊 Dimensões: {df.shape[0]} linhas × {df.shape[1]} colunas")
````

## Limpeza e Preparação de Dados

Na vida real, os dados raramente vêm limpos e prontos para análise. Vamos simular alguns problemas comuns e aprender a resolvê-los.

### Problemas Comuns:
- **Valores faltantes** (NaN, null)
- **Dados duplicados**
- **Outliers** (valores extremos)
- **Inconsistências** nos dados

### Estratégias de Tratamento:
1. **Remoção**: Deletar linhas/colunas problemáticas
2. **Imputação**: Preencher com valores calculados

````python
# Simulando problemas nos dados para fins educativos
print("SIMULANDO PROBLEMAS NOS DADOS:")
print("=" * 40)

# Criando uma cópia para experimentar
df_dirty = df.copy()

# Introduzindo valores faltantes aleatoriamente
np.random.seed(42)  # Para reprodutibilidade
for col in df_dirty.columns[:-1]:  # Não mexe na coluna species
    # Seleciona 5 posições aleatórias para cada coluna
    missing_indices = np.random.choice(df_dirty.index, 5, replace=False)
    df_dirty.loc[missing_indices, col] = np.nan

print("Valores faltantes inseridos!")
print(f"Total de valores faltantes: {df_dirty.isnull().sum().sum()}")
````

## Estratégias de Limpeza

Vamos explorar diferentes abordagens para lidar com valores faltantes:

### Estratégia 1: Remoção com dropna()
- **Vantagem**: Simples e rápida
- **Desvantagem**: Perde informação
- **Quando usar**: Poucos valores faltantes (< 5%)




````python
# Estratégia 1: Remoção de linhas com valores faltantes

df_dropped = df_dirty.dropna()

display(df_dirty.info(),df_dropped.info())
````

### Estratégia 2: Imputação com fillna()
- **Média**: Para dados numéricos sem outliers
- **Mediana**: Para dados com outliers
- **Moda**: Para dados categóricos
- **Interpolação**: Para séries temporais


````python
# Estratégia 2: Imputação com média


df_imputed = df_dirty.copy()
numeric_columns = df_imputed.select_dtypes(include=[np.number]).columns

# Calcula médias
mean_values = df_imputed[numeric_columns].mean()

# Conta valores ausentes
missing_counts = df_imputed[numeric_columns].isnull().sum()

# Preenche os valores
df_imputed.fillna(mean_values, inplace=True)

# Imprime resumo
for col in numeric_columns:
    print(f"{col}: {missing_counts[col]} valores preenchidos com {mean_values[col]:.2f}")

# Verifica resultado
print(f"\nValores faltantes após imputação: {df_imputed.isnull().sum().sum()}")

display(df_dirty.info(), df_imputed.info())
````

- **Outliers**: Valores extremos isolados?

###  Scatter Plots: Descobrindo Relacionamentos

Os **Scatter Plots** revelam relacionamentos entre variáveis. São fundamentais para entender correlações e padrões.

### Padrões a observar:
- **Linear**: Pontos seguem uma linha reta
- **Curvilíneo**: Relação não-linear
- **Clusters**: Grupos distintos de pontos
- **Outliers**: Pontos isolados


````python
# Scatter plot básico
# Grafico de dispersão

plt.scatter(df['sepal_length'], df['petal_length'])

plt.title('Sepal Length vs Petal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.show()
````

````python
# Os mesmos dados mas agora cada classe de uma cor diferente

colors = {'Iris-setosa':'red', 'Iris-versicolor':'blue', 'Iris-virginica':'green'}

plt.scatter(df['sepal_length'], df['petal_length'], c=df['species'].map(colors), label=colors)

plt.title('Sepal Length vs Petal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.legend(handles=[plt.Line2D([0], [0], marker='o', color='w', markerfacecolor=color, markersize=10) for color in colors.values()], labels=colors.keys())
plt.show()
````

## Análise de Correlação

A **correlação** mede a força e direção do relacionamento linear entre variáveis.

### 📏 Escala de Interpretação:
| Valor | Interpretação |
|-------|---------------|
| 0.9 a 1.0 | Correlação muito forte |
| 0.7 a 0.9 | Correlação forte |
| 0.5 a 0.7 | Correlação moderada |
| 0.3 a 0.5 | Correlação fraca |
| 0.0 a 0.3 | Correlação desprezível |

### Importante:

**Correlação ≠ Causalidade**
- Uma correlação alta não significa que uma variável causa a outra
- Pode haver uma terceira variável influenciando ambas


````python
cols = ['sepal_length', 'sepal_width', 'petal_length','petal_width']
corr_matx = df[cols].corr()

heatmap = sns.heatmap(corr_matx,cbar=True,annot=True,square=True,fmt='.2f',annot_kws={'size': 15},yticklabels=cols,xticklabels=cols,cmap='Dark2')
````
